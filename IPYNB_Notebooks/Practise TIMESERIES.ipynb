## Bulding and Evaluating Time Series Forecast

```python
##Splitting  time series intotraining nad test sets
```

```python
len(data)
```

```python
train_len=120
train=data[0:train_len]
test=data[train_len:]
```

## NAIVE METHOD

```python
y_hat_naive=test.copy()
```

```python
train['Passengers'][train_len-1]
```

```python
y_hat_naive['naive_forecast']=train['Passengers'][train_len-1]
```

### Ploting train , test and forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['Passengers'],label='Train')
plt.plot(test['Passengers'],label='Test')
plt.plot(y_hat_naive['naive_forecast'],label='Naive Forecast')
plt.legend(loc='best')
plt.title('Naive Method')
plt.show()
```

## Calculate RMSE and MAPE

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse=np.sqrt(mean_squared_error(test['Passengers'],y_hat_naive['naive_forecast'])).round(2)
mape=np.round(np.mean(np.abs(test['Passengers']-y_hat_naive['naive_forecast'])/test['Passengers'])*100,2)
result=pd.DataFrame({'Method':['Naive Method'],'MAPE':[mape],'RMSE':[rmse]})
result=result[['Method','RMSE','MAPE']]
result
```

### Simple Average Method

```python
y_hat_avg=test.copy()
y_hat_avg['avg_forecast']=train['Passengers'].mean()
```

### Plotting the train test forecast

```python
plt.figure(figsize=[12,4])
plt.plot(train['Passengers'],label='Train')
plt.plot(test['Passengers'],label='Test')
plt.plot(y_hat_avg['avg_forecast'],label='Avg_Forecast')
plt.legend()
plt.title('AVERAGE FORECAST')
plt.show()
```

```python
rmse=np.sqrt(mean_squared_error(test['Passengers'],y_hat_avg['avg_forecast'])).round(2)
mape=np.round(np.mean(np.abs(test['Passengers']-y_hat_avg['avg_forecast'])/test['Passengers'])*100,2)

mape = np.round(np.mean(np.abs(test['Passengers']-y_hat_avg['avg_forecast'])/test['Passengers'])*100,2)

result1=pd.DataFrame({'Method':'AVERAGE','RMSE':[rmse],'MAPE':[mape]})
result=pd.concat([result,result1])
result
```

## Simple Moving Average method

```python
y_hat_sma=data.copy()
ma_window=3
y_hat_sma['sma_forecast']=data['Passengers'].rolling(ma_window).mean()
y_hat_sma['sma_forecast'][train_len:]=y_hat_sma['sma_forecast'][train_len-1]
```

## Calculate RMSE and MAPE

```python
rmse=np.sqrt(mean_squared_error(test['Passengers'],y_hat_sma['sma_forecast'][train_len:])).round(2)
mape=np.round(np.mean(np.abs(test['Passengers']-y_hat_sma['sma_forecast'][train_len:])/test['Passengers'])*100,2)

tempresult=pd.DataFrame({'Method':'Simple Average Moving_window3','RMSE':[rmse],'MAPE':[mape]})
result=pd.concat([result,tempresult])
result
```

## Holt's exponential smoothing method - TREND

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model=ExponentialSmoothing(np.array(train['Passengers']),seasonal_periods=12,trend='additive',seasonal=None)
model_fit=model.fit(smoothing_level=0.2,smoothing_slope=0.01,optimized=False)
print(model_fit.params)
```

```python
y_hat_holt=test.copy()
y_hat_holt['holt_forecast']=model_fit.forecast(24)
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['Passengers'],label='Train')
plt.plot(test['Passengers'],label='Test')
plt.plot(y_hat_holt['holt_forecast'],label='Holts Exponential Simple Method')
plt.legend()
plt.title('Holts Expnential Method')
plt.show()
```

## RMSE and MAPE

```python
rmse=np.sqrt(mean_squared_error(test['Passengers'],y_hat_holt['holt_forecast'])).round(2)
mape=np.round(np.mean(np.abs(test['Passengers']-y_hat_holt['holt_forecast'])/test['Passengers'])*100,2)

tempresult=pd.DataFrame({'Method':'HOLTs Exponential smoothing0.4 ','RMSE':[rmse],'MAPE':[mape]})
result=pd.concat([result,tempresult])
result
```
