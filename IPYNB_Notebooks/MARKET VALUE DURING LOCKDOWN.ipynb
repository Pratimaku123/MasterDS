```python
from sklearn import model_selection
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,train_size=0.8,random_state=100,stratify=y,shuffle=True)
```

```python
import xgboost as xgb
from sklearn import metrics 
xgclf=xgb.XGBClassifier()
xgclf.fit(X_train,y_train)
```

```python
print('AUC on train data by XGBOOST',metrics.roc_auc_score(y_true=y_train,y_score=xgclf.predict_proba(X_train)[:,1]))
```

```python
print('AUC on test data by XGBOOST',metrics.roc_auc_score(y_true=y_test,y_score=xgclf.predict_proba(X_test)[:,1]))
```

```python
xgb_model=xgb.XGBClassifier()
parameters={
    'learning_rate':[0.3],
    'max_depth':[6],
    'min_child_weight':[1],
    'n_estimators':[100]
}

scorer=metrics.make_scorer(metrics.roc_auc_score,greater_is_better=True,needs_proba=True,needs_threshold=False)

clf_xgb=model_selection.GridSearchCV(estimator=xgb_model,param_grid=parameters,n_jobs=-1,cv=3,scoring=scorer,refit=True)
clf_xgb.fit(X_train,y_train)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
dfs = []
for dirname, _, filenames in os.walk('/home/pkumari/Downloads/archive/'):
    for filename in filenames:
        dfs.append(pd.read_csv(os.path.join(dirname, filename)))
```

```python
dfs_all=pd.concat(dfs,axis=0)
```

```python
dfs_all['date_1']=pd.to_datetime(dfs_all['<date>'],format='%Y%m%d',errors='ignore')
```

```python
dfs_all
```

```python
def plot_share_history(dfplot, tickername, draw_lockdown_lines=["2020-03-25", "2020-04-15","2020-05-04","2020-05-18"]):
    dfplot=dfs_all[["date_1","<close>"]][df_all["<ticker>"] == tickername]
    dfplot=dfplot.set_index("date_1")
    ax=dfplot.plot(title="Share Price history - {}".format(tickername))
    if draw_lockdown_lines is not None:
        for lock_date in draw_lockdown_lines:
            ax.axvline(lock_date, color="red", linestyle="--")

plot_share_history(dfs_all, "AFFLE")
plot_share_history(dfs_all, "ALMONDZ")
plot_share_history(dfs_all, "BALKRISHNA")
plot_share_history(dfs_all, "SIEMENS")
```

### Bulid and Evaluate times series forecaasting

```python
import math
```

```python
train_len=math.ceil(0.6*len(dfplot))
```

```python
train=dfplot[0:train_len]
test=dfplot[train_len:]
```

```python
len(dfplot)
```

```python
len(test)
```

## NAIVE METHOD

```python
y_hat_naive=test.copy()
```

```python
y_hat_naive['Niave_forecast']=train['<close>'][train_len-1]
```

## plot train tets and  NAIVE forecast

```python
plt.figure(figsize=[12,4])
plt.plot(train['<close>'],label='TRAIN')
plt.plot(test['<close>'],label='TEST')
plt.plot(y_hat_naive['Niave_forecast'],label='Naive Forecast')
plt.legend(loc='best')
plt.title('NAIVE METHOD for GRINDWELL STOCK')
plt.show()
```

## Caculate RMSE and MAPE 

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse=np.sqrt(mean_squared_error(test['<close>'],y_hat_naive['Niave_forecast'])).round(2)
mape=np.round(np.mean(np.abs(test['<close>']-y_hat_naive['Niave_forecast'])/test['<close>'])*100,2)
result=pd.DataFrame({'Method':'Naive','RMSE':[rmse],'MAPE':[mape]})
result
```

## SImple Average Method

```python
y_hat_avg=test.copy()
```

```python
y_hat_avg['avg_forecast']=train['<close>'].mean()
```

## Plotting train test forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['<close>'],label='Train')
plt.plot(test['<close>'],label='Test')
plt.plot(y_hat_avg['avg_forecast'],label='Avg_Forcast')
plt.legend()
plt.title('AVERAGE_FORECAST')
plt.show()
```

```python
rmse=np.sqrt(mean_squared_error(test['<close>'],y_hat_avg['avg_forecast'])).round(2)
mape=np.round(np.mean(np.abs(test['<close>']-y_hat_avg['avg_forecast'])/test['<close>'])*100,2)

result=pd.DataFrame({'Method':'AVERAGE','RMSE':[rmse],'MAPE':[mape]})
result
```
